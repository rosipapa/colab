## Questions

```python
# Q1. When are shifts most frequently left unfilled?
def shifts_nao_preenchidos_por_dia_hora(df):
    df_unfilled = df[df['CLAIMED_AT'].isnull()]
    pivot = pd.pivot_table(df_unfilled, index='SHIFT_WEEKDAY', columns='SHIFT_HOUR', values='SHIFT_ID', aggfunc='count', fill_value=0)
    pivot = pivot.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
    plt.figure(figsize=(15, 6))
    sns.heatmap(pivot, cmap="Reds", annot=True, fmt="d")
    plt.title("Turnos não preenchidos por dia da semana e hora")
    plt.ylabel("Dia da Semana")
    plt.xlabel("Hora do Turno")
    plt.show()

shifts_nao_preenchidos_por_dia_hora(df)
```

```python
# Q2. When are shifts most frequently left unfilled?

def taxa_de_preenchimento(df):
    taxa = df['CLAIMED'].mean()
    print(f"✅ Taxa de turnos preenchidos: {taxa:.2%}")

taxa_de_preenchimento(df)
```

```python
# Q3. What is the distribution of shifts by time of day (SLOT)?
def distribuicao_por_horario(df):
    sns.countplot(data=df, x='SLOT', hue='CLAIMED')
    plt.title("Distribuição de turnos por período (am/pm) e preenchimento")
    plt.show()
distribuicao_por_horario(df)
```

```python
# Q4. Is there a relationship between PAY_RATE and shift fulfillment?
def relacao_pagamento_preenchimento(df):
    sns.boxplot(x='CLAIMED', y='PAY_RATE', data=df)
    plt.title("Relação entre PAY_RATE e preenchimento do turno")
    plt.xticks([0,1], ['Não preenchido', 'Preenchido'])
    plt.show()

relacao_pagamento_preenchimento(df)
```

```python
# Q5. How does lead time affect fulfillment?
def antecedencia_criacao(df):
    df['ANTECEDENCIA_HRS'] = (df['SHIFT_START_AT'] - df['SHIFT_CREATED_AT']).dt.total_seconds() / 3600
    sns.histplot(data=df, x='ANTECEDENCIA_HRS', hue='CLAIMED', bins=30, kde=True)
    plt.title("Antecedência da criação dos turnos (em horas)")
    plt.xlim(0, 100)
    plt.show()

antecedencia_criacao(df)
```

```python
# Q8. Does delay in first view affect fill rate by SLOT


def tempo_visualizacao_por_slot(df):
    df["tempo_view"] = (df["OFFER_VIEWED_AT"] - df["SHIFT_CREATED_AT"]).dt.total_seconds() / 3600

    # Agrupamento por SLOT
    resumo = df.groupby("SLOT").agg(
        tempo_medio_visualizacao=("tempo_view", "mean"),
        taxa_nao_preenchimento=("IS_FILLED", lambda x: 1 - x.mean()),
        n_turnos=("SHIFT_ID", "count")
    ).reset_index()

    print("🔍 Análise de tempo médio de visualização vs. taxa de não preenchimento por SLOT:")
    print(resumo)

    # Visualização
    ax = sns.scatterplot(
        data=resumo,
        x="tempo_medio_visualizacao",
        y="taxa_nao_preenchimento",
        size="n_turnos",
        hue="SLOT",
        sizes=(50, 500),
        legend="full"
    )
    plt.title("Correlação entre tempo até visualização e não preenchimento (por SLOT)")
    plt.xlabel("Tempo médio até visualização (h)")
    plt.ylabel("Taxa de turnos não preenchidos")
    plt.show()
```

```python
tempo_visualizacao_por_slot(df)
```

```python
# Q7. Does lead time affect shift fill rates?

def taxa_preenchimento_vs_antecedencia(df):
    df["antecedencia_criacao_h"] = (df["SHIFT_START_AT"] - df["SHIFT_CREATED_AT"]).dt.total_seconds() / 3600

    # Categorizar em bins
    bins = [0, 6, 24, 72, 168, 720]
    labels = ["<6h", "6-24h", "1-3d", "3-7d", ">7d"]
    df["grupo_antecedencia"] = pd.cut(df["antecedencia_criacao_h"], bins=bins, labels=labels)

    # Agrupamento
    resultado = df.groupby("grupo_antecedencia").agg(
        taxa_preenchimento=("IS_FILLED", "mean"),
        n_turnos=("SHIFT_ID", "count")
    ).reset_index()

    print("📈 Taxa de preenchimento por antecedência de criação do turno:")
    print(resultado)

    # Visualização
    sns.barplot(data=resultado, x="grupo_antecedencia", y="taxa_preenchimento", palette="viridis")
    plt.title("Taxa de preenchimento por faixa de antecedência (horas)")
    plt.ylabel("Taxa de preenchimento")
    plt.xlabel("Antecedência da criação do turno")
    plt.ylim(0, 1)
    plt.show()
```

```python
taxa_preenchimento_vs_antecedencia(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway, kruskal
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# Feature Engineering
df["IS_FILLED"] = df["CLAIMED_AT"].notna()
df["tempo_view"] = (df["OFFER_VIEWED_AT"] - df["SHIFT_CREATED_AT"]).dt.total_seconds() / 3600
df["antecedencia_criacao_h"] = (df["SHIFT_START_AT"] - df["SHIFT_CREATED_AT"]).dt.total_seconds() / 3600
df["grupo_antecedencia"] = pd.cut(df["antecedencia_criacao_h"],
                                  bins=[0, 6, 24, 72, 168, 720],
                                  labels=["<6h", "6-24h", "1-3d", "3-7d", ">7d"])
```

```python
df.columns
```

```python
# Q6. Is there a statistically significant relationship between lead time and fill rate?

anova_data = df.dropna(subset=["grupo_antecedencia", "IS_FILLED"])
groups = [group["IS_FILLED"].astype(int).values for name, group in anova_data.groupby("grupo_antecedencia")]

anova_result = f_oneway(*groups)
kruskal_result = kruskal(*groups)

print("ANOVA result:", anova_result)
print("Kruskal-Wallis result:", kruskal_result)
```

```python
# Q9. Can we predict whether a shift will be filled based on its characteristics?

df_model = df[["DURATION", "PAY_RATE", "CHARGE_RATE", "tempo_view", "antecedencia_criacao_h", "IS_FILLED"]].dropna()

X = df_model.drop("IS_FILLED", axis=1)
y = df_model["IS_FILLED"].astype(int)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print("Feature Importances:")
for feature, importance in zip(X.columns, rf.feature_importances_):
    print(f"{feature}: {importance:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
